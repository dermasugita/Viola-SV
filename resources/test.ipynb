```python
bedpe_detail = bedpe_test.filter(['svtype == DUP', 'svtype == DEL'], query_logic='or')
```

```python
def func(f):
    if f.startswith('.'):
        return
    file = os.path.join('pcawg', f)
    bedpe = sv_parser.read_bedpe(file, svtype_col_name='svclass')
    bedpe = bedpe.filter(['svtype == DUP', 'svtype == DEL'], query_logic='or')
    bedpe = bedpe.get_detail_svtype().to_bedpe_like(how='expand', custom_infonames=['svtype', 'svtype_detail'])
    bedpe['sample'] = f
    return bedpe
```

```python
all_bedpe = [func(f) for f in os.listdir('pcawg')]
```

```python
df_all_bedpe = pd.concat(all_bedpe)
```

```python
df_count_all = df_all_bedpe.pivot_table(index='sample', columns='svtype_detail_0', aggfunc={'name': 'count'})
```
